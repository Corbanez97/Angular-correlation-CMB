```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T=np.linspace(10e-2,0.6,1000)
X=np.empty(len(T))

for i in range (len(T)):
    c=(2e-9)*1.2*(2*np.pi*T[i]/(0.5e6))**(3/2)*np.exp(13.6/T[i])/(np.pi**2)
    X[i]=np.roots([c,1,-1])[1]
 

plt.rcParams['figure.figsize']=[10,8]
plt.rcParams.update({'font.size':18})

plt.xlabel('Temperatura(eV)')
plt.ylabel('$X_e$')
plt.plot(T,X)
```

```python
Temp=np.linspace(0.3,0.35,1000)
Xe=np.empty(len(T))

for i in range (len(Temp)):
    ce=(2e-9)*1.2*(2*np.pi*Temp[i]/(0.5e6))**(3/2)*np.exp(13.6/Temp[i])/(np.pi**2)
    Xe[i]=np.roots([ce,1,-1])[1]
 

plt.rcParams['figure.figsize']=[10,8]
plt.rcParams.update({'font.size':18})

plt.xlabel('Temperatura(eV)')
plt.ylabel('$X_e$')
plt.plot(Temp,Xe)
```

```python
z=np.linspace(200,1800,1000)
Xee=np.empty(len(z))

for i in range (len(z)):
    cz=(2e-9)*1.2*(2*np.pi*(6.6e-4)*(1+z[i])/0.5e6)**(3/2)*np.exp(13.6/((6.6e-4)*(1+z[i])))/(np.pi**2)
    Xee[i]=np.roots([cz,1,-1])[1]
 

plt.rcParams['figure.figsize']=[10,8]
plt.rcParams.update({'font.size':18})

plt.xlabel('z')
plt.ylabel('$X_e$')
plt.plot(z,Xee)
```
